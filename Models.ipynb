```python
import pandas as pd
add = 'RSI'
```

```python
import warnings
# Suppress the specific FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn.ensemble._weight_boosting')
```

```python
df = pd.read_csv('Featured_Engineering_Data_To_CSV.csv')
print(df.shape)
df.head(30)
```

```python
df.date = pd.to_datetime(df.date)
df.set_index('date',inplace=True)
df.info()
```

```python
df["Label2"] = df.groupby("Name").Pct_Change.shift(-1)
df["Label"] = df["Label2"] > 0
df.drop(columns=['Label2'], inplace=True)
```

```python
### import models from sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

models = {

    'Random Forest': RandomForestClassifier(n_estimators=100,max_depth=1),
    'AdaBoost': AdaBoostClassifier(n_estimators=100),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),

}
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# בונה רשימה של המניות ללא כפיליות
stocks = df.Name.unique().tolist()

# הגדרת הפיטצ'רים שלי
features_list = ['open', 'high', 'low', 'close', add]  
all_accuracies = {}  

# הנתונים שאני רוצה שהוא יתאמן עליהם
dff = df[features_list + ['Name', 'Label']].copy()
dff.dropna(inplace=True)

# לולאה על כל מניה
for stock in stocks:
    #הגדרת הנתונים בשביל מניה ספציפית 
    data = dff[dff.Name == stock].copy()
    
    # הגדרת X והגדרת Y
    X = data[features_list]
    y = data['Label']
    
    # הגדרת הטריין והטסט שלי
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    # מילון לשמירת דיוק המודלים עבור המניה הנוכחית
    accuracy_results = {}
    
    # לולאה על כל מודל
    for name, model in models.items():

         # אימון המודל
        model.fit(X_train, y_train)
            
        #תוצאת הטסט שלי
        y_pred = model.predict(X_test)
            
        # חישוב הדיוק של המודל
        accuracy = accuracy_score(y_test, y_pred)
            
         # שמירת הדיוק של המודל במילון
        accuracy_results[name] = accuracy
        
    
    # שמירת דיוק המודלים עבור המניה הנוכחית
    all_accuracies[stock] = accuracy_results
    print(stock)  #הדפסה בשביל לראות התקדמות 

# הדפסת המילון לדאטה פריים שיהיה יותר נוח
accuracy_df = pd.DataFrame(all_accuracies).T

# הצגת חמש השורות הראשונות של התוצאות
accuracy_df.head()
```

```python
accuracy_df.to_csv(f"stocks_model_accuracies_base+{add}.csv")
```
