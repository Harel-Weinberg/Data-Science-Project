```python
import pandas as pd
import os
```

```python
all_years = pd.read_csv("all_stocks_5yr.csv")
row_counts = all_years.groupby(all_years.Name).size()
stocks_with_few_rows = row_counts[row_counts < 500].index
stocks_with_few_rows
```

####################### VALUATION OF LDA #######################

```python
# הגדרת התיקייה שבה נמצאים הקבצים
folder_path = 'LDA'

# מציאת כל הקבצים שמסתיימים ב-csv בתיקייה
csv_files = [f for f in os.listdir(folder_path) if (f.endswith('.csv'))]

# מוציא את הרשימת מניות שלי
temp_df = pd.read_csv(os.path.join(folder_path, csv_files[0]))
temp_df = temp_df[~temp_df["Unnamed: 0"].isin(stocks_with_few_rows)]
print(temp_df.head())
```

```python
# המרה לדאטה פריים
LinearDiscriminantAnalysis = pd.DataFrame()
LinearDiscriminantAnalysis["Name"] = temp_df["Unnamed: 0"]
```

```python
for file in csv_files:
    file_path = os.path.join(folder_path, file)

    df = pd.read_csv(file_path)
    df = df[~df["Unnamed: 0"].isin(stocks_with_few_rows)] # מוריד את כל המניות עם פחות מ500 שורות
    LinearDiscriminantAnalysis[file[:-4]] = df["LDA"]
```

```python
LinearDiscriminantAnalysis[LinearDiscriminantAnalysis["Name"] == "APTV"]
```

```python
LinearDiscriminantAnalysis.head(3)
```

```python
LinearDiscriminantAnalysis.set_index('Name', inplace=True)

# מציאת הערך הכי גבוה
max_values = LinearDiscriminantAnalysis.max()
# מציאת האינדקס של הערך ממקס שסימנו מקודם
max_indices = LinearDiscriminantAnalysis.idxmax()

# יצירת DataFrame חדש עם התוצאות
result = pd.DataFrame({
    'Column': max_values.index,
    'Max Value': max_values.values,
    'Stock Name': max_indices.values
})
result.sort_values(ascending=False, by="Max Value")
```

```python
LinearDiscriminantAnalysis.describe()
# חישוב עמודות הממוצע
mean_values = LinearDiscriminantAnalysis.mean()
# חישוב הממוצע של הממוצעים
mean_of_means = mean_values.mean()
print(mean_of_means)
```
