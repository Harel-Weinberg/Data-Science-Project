```python
import pandas as pd
add = ['open' , 'high', 'low' ,	'close' ,'volume',]
```

```python
df = pd.read_csv('Featured_Engineering_Data_To_CSV.csv')
```

```python
df["Label2"] = df.groupby("Name").Pct_Change.shift(-1)
df["Label"] = df["Label2"] > 0
df.drop(columns=['Label2'], inplace=True)
```

```python
# בחירת הפיטצ'רים שלי
stocks = df.Name.unique().tolist()
features_list = add
all_accuracies = {}
dff = df[features_list+['Name','Label']].copy()
dff.dropna(inplace=True)
accuracy_results = {}
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


# יצירת מודל LDA
model = LinearDiscriminantAnalysis()
for stock in stocks:
    print(stock)

    data = dff[dff.Name == stock].copy()
    X = data[features_list]
    y = data['Label']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# התאמת המודל לנתוני האימון
    model.fit(X_train, y_train)

# ביצוע ניבוי על נתוני הבדיקה
    y_pred = model.predict(X_test)

# הערכת המודל
    accuracy = accuracy_score(y_test, y_pred)
# making df
    accuracy_results = {"LDA": accuracy}
    
    # עדכון התוצאות לכל מניה
    all_accuracies[stock] = accuracy_results
accuracy_df = pd.DataFrame(all_accuracies).T


accuracy_df.head()
```

```python
accuracy_df.to_csv(f"LDA+{add}.csv")
```
